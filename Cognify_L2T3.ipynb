```python
import pandas as pd

# Load the dataset (replace 'your_dataset.csv' with the actual dataset file path)
df = pd.read_csv('/content/Dataset .csv')

# Task 6.1: Extract additional features such as the length of the restaurant name or address
# Assuming 'Restaurant Name' and 'Address' columns exist
if 'Restaurant Name' in df.columns:
    df['Name Length'] = df['Restaurant Name'].apply(len)  # Length of the restaurant name

if 'Address' in df.columns:
    df['Address Length'] = df['Address'].apply(len)  # Length of the restaurant address

# Display the first few rows to check the new features
print("\nData with extracted features:")
print(df[['Restaurant Name', 'Address', 'Name Length', 'Address Length']].head())

# Task 6.2: Create new features like "Has Table Booking" or "Has Online Delivery" by encoding categorical variables
# Assuming 'Table booking' and 'Online delivery' columns are categorical (True/False or 1/0)

# Create a binary feature for table booking
booking_column = next((col for col in df.columns if col.lower().strip() == 'table booking'), None)

# If the column is found, proceed with creating the binary feature
if booking_column:
    df['Has Table Booking'] = df[booking_column].apply(lambda x: 1 if x else 0)
else:
    print("Warning: 'Table booking' column not found. Skipping 'Has Table Booking' feature creation.")

# Similarly, for 'Online delivery' column:
delivery_column = next((col for col in df.columns if 'online delivery' in col.lower()), None)

if delivery_column:
    df['Has Online Delivery'] = df[delivery_column].apply(lambda x: 1 if x else 0)
else:
    print("Warning: 'Online delivery' column not found. Skipping 'Has Online Delivery' feature creation.")

# Create a binary feature for online delivery
if delivery_column:
    df['Has Online Delivery'] = df[delivery_column].apply(lambda x: 1 if x else 0)
else:
    print("Warning: 'Online delivery' column not found. Skipping 'Has Online Delivery' feature creation.")

# Display the new binary features
print("\nData with binary features for Table Booking and Online Delivery:")

# Check if the columns exist before accessing them
columns_to_display = ['Has Table Booking', 'Has Online Delivery']
for col in ['Table booking', 'Online delivery']:
    if col in df.columns:
        columns_to_display.insert(0, col)  # Add original column at the beginning

# Display only available columns
print(df[[col for col in columns_to_display if col in df.columns]].head())

# Task 6.3: Optionally, create additional features (example: Extracting 'City' from 'Address' if 'City' isn't available)
# Assuming 'City' is not directly available, you might extract it from the 'Address' column using some logic.
# Here, let's assume the 'Address' column contains the city at the end, separated by commas.
columns_to_display = ['Has Table Booking', 'Has Online Delivery']

# Add any original columns if they exist
for original_col in ['Table booking', 'Online delivery']:
    if original_col in df.columns:
        columns_to_display.insert(0, original_col)  # Add original column at the beginning

# Display only available columns
print("\nData with binary features for Table Booking and Online Delivery:")
print(df[[col for col in columns_to_display if col in df.columns]].head())
if 'Address' in df.columns:
    df['City Extracted'] = df['Address'].apply(lambda x: x.split(',')[-2] if len(x.split(',')) > 1 else 'Unknown')

# Display the data with the extracted city feature
print("\nData with extracted 'City' feature from Address:")
print(df[['Address', 'City Extracted']].head())

```
