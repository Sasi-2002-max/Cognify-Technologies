```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (replace 'your_dataset.csv' with the actual dataset file path)
df = pd.read_csv('/content/Dataset .csv')

# Task 4.1: Determine the percentage of restaurants that offer table booking and online delivery
# Assuming 'Table booking' and 'Online delivery' are columns indicating availability (True/False or 1/0)
# Check for the unique values in these columns to understand how they are stored

print(df.columns)  # Print columns to verify name

# Task 4.1: Determine the percentage of restaurants that offer table booking and online delivery

print("\nUnique values for 'Table booking' column:")
print(df['Has Table booking'].value_counts())  # Changed column name to 'Has Table booking'

# Similarly, for 'Online delivery', verify the column name and adjust accordingly
print("\nUnique values for 'Online delivery' column:")
print(df['Has Online delivery'].value_counts())

# Task 4.2: Compare the average ratings of restaurants with table booking and those without
# First, let's calculate the average rating for restaurants with and without table booking
table_booking_ratings = df[df['Has Table booking'] == 'Yes']['Aggregate rating'] # Changed column name to 'Has Table booking' and compared with 'Yes'
no_table_booking_ratings = df[df['Has Table booking'] == 'No']['Aggregate rating'] # Changed column name to 'Has Table booking' and compared with 'No'


# Calculate mean ratings
mean_rating_with_booking = table_booking_ratings.mean()
mean_rating_without_booking = no_table_booking_ratings.mean()

print(f"\nAverage rating of restaurants with table booking: {mean_rating_with_booking:.2f}")
print(f"Average rating of restaurants without table booking: {mean_rating_without_booking:.2f}")

# Visualize the ratings comparison
plt.figure(figsize=(8, 6))
sns.boxplot(x='Has Table booking', y='Aggregate rating', data=df) # Changed column name to 'Has Table booking'
plt.title('Comparison of Ratings: Table Booking vs. No Table Booking')
plt.xlabel('Table Booking')
plt.ylabel('Aggregate Rating')
plt.show()
# Task 4.3: Analyze the availability of online delivery
# Analyze the distribution of online delivery (True/False or 1/0) and its impact on the ratings

# First, let's calculate the average rating for restaurants with and without online delivery
online_delivery_ratings = df[df['Has Online delivery'] == 'Yes']['Aggregate rating'] # Changed column name to 'Has Online delivery' and compared with 'Yes'
no_online_delivery_ratings = df[df['Has Online delivery'] == 'No']['Aggregate rating'] # Changed column name to 'Has Online delivery' and compared with 'No'

# Calculate mean ratings
mean_rating_with_delivery = online_delivery_ratings.mean()
mean_rating_without_delivery = no_online_delivery_ratings.mean()

print(f"\nAverage rating of restaurants with online delivery: {mean_rating_with_delivery:.2f}")
print(f"Average rating of restaurants without online delivery: {mean_rating_without_delivery:.2f}")

# Visualize the ratings comparison for online delivery
plt.figure(figsize=(8, 6))
sns.boxplot(x='Has Online delivery', y='Aggregate rating', data=df) # Changed column name to 'Has Online delivery'
plt.title('Comparison of Ratings: Online Delivery vs. No Online Delivery')
plt.xlabel('Online Delivery')
plt.ylabel('Aggregate Rating')
plt.show()
```
