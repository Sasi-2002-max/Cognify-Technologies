```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (replace 'your_dataset.csv' with the actual dataset file path)
df = pd.read_csv('/content/Dataset .csv')

# Task 5.1: Determine the most common price range among all the restaurants
# Assuming the 'Price range' column contains the price category (e.g., 1-4)
price_range_counts = df['Price range'].value_counts()

# Display the most common price range
most_common_price_range = price_range_counts.idxmax()
print(f"\nMost common price range: {most_common_price_range}")
print(f"Number of restaurants in this price range: {price_range_counts[most_common_price_range]}")

# Task 5.2: Calculate the average rating for each price range
# Group by the 'Price range' and calculate the average of the 'Aggregate rating'
average_rating_by_price_range = df.groupby('Price range')['Aggregate rating'].mean()

# Display the average ratings by price range
print("\nAverage ratings for each price range:")
print(average_rating_by_price_range)

# Task 5.3: Identify the color that represents the highest average rating among different price ranges
# Let's visualize the average ratings across price ranges with a color-coded bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=average_rating_by_price_range.index, y=average_rating_by_price_range.values, palette='viridis')
plt.title('Average Rating for Each Price Range')
plt.xlabel('Price Range')
plt.ylabel('Average Rating')
plt.show()

# Identify the price range with the highest average rating
highest_avg_rating_price_range = average_rating_by_price_range.idxmax()
highest_avg_rating_value = average_rating_by_price_range.max()

print(f"\nPrice range with the highest average rating: {highest_avg_rating_price_range}")
print(f"Highest average rating value: {highest_avg_rating_value:.2f}")

# We can also print the corresponding color for the highest rating
# Using the same palette, get the color that corresponds to the highest rating
palette = sns.color_palette('viridis')
color_for_highest_rating = palette[int((highest_avg_rating_value / average_rating_by_price_range.max()) * (len(palette) - 1))]

print(f"Color representing the highest rating: {color_for_highest_rating}")
```
