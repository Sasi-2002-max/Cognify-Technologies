```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (replace 'your_dataset.csv' with the actual dataset file path)
df = pd.read_csv('/content/Dataset .csv')

# Task 2.1: Calculate basic statistical measures for numerical columns
# This includes mean, median, standard deviation, etc.
print("\nBasic statistical measures for numerical columns:")
print(df.describe())  # This provides a summary of the statistics for numerical columns

# Task 2.2: Explore the distribution of categorical variables (Country Code, City, Cuisines)

# Distribution of Country Code
plt.figure(figsize=(8, 6))
sns.countplot(x='Country Code', data=df)
plt.title('Distribution of Country Code')
plt.xlabel('Country Code')
plt.ylabel('Count')
plt.show()

# Distribution of City (Note: This can be large, so consider showing top cities)
top_cities = df['City'].value_counts().nlargest(10)
plt.figure(figsize=(8, 6))
sns.barplot(x=top_cities.index, y=top_cities.values)
plt.title('Top 10 Cities with Highest Number of Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()

# Distribution of Cuisines (If multiple cuisines are listed in one column, split them)
# First, handle cases where 'Cuisines' column may have missing values or be in a list format.
df['Cuisines'] = df['Cuisines'].fillna('Unknown')  # Fill missing cuisines with 'Unknown'

# Split cuisines if they are comma-separated
cuisines_split = df['Cuisines'].str.split(',', expand=True).stack()
top_cuisines = cuisines_split.value_counts().nlargest(10)

# Plot the top 10 cuisines with the highest number of restaurants
plt.figure(figsize=(8, 6))
sns.barplot(x=top_cuisines.index, y=top_cuisines.values)
plt.title('Top 10 Cuisines with Highest Number of Restaurants')
plt.xlabel('Cuisines')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()

# Task 2.3: Identify the top cuisines and cities with the highest number of restaurants

# Top cuisines
print("\nTop cuisines with the highest number of restaurants:")
print(top_cuisines)

# Top cities
print("\nTop cities with the highest number of restaurants:")
print(top_cities)
```
